```python
target_name = "MANZANO"
SEED_NUM = "1"
redacted_name = ""
vowels = "AEIOU"

print("EXERCISE 1: REDACTION")
print(f"Original Surname: {target_name}")
print(f"SEED_NUM Used: {SEED_NUM}")
print("\nExecution Log:")

for letters in target_name:
    if letters in vowels:
        redacted_name = (redacted_name + SEED_NUM)
        print(f"  '{letters}' is vowel → replaced with {SEED_NUM}")
    else:
        redacted_name = (redacted_name + letters)
        print(f"  '{letters}' is consonant → kept as '{letters}'")
print(f"\nRedacted Surname: {redacted_name}")
```

```python
first_letter = target_name[0]
ascii_value = ord(first_letter)
seed_num = 1
loops = 1 + 3

# Calculate verification hash
verification_hash = (ascii_value * seed_num) - loops

print("EXERCISE 2: Forensic Signature")
print(f"    Engineer Name: {target_name}")
print(f"    First Letter: '{first_letter}'")
print(f"    ASCII Value (A): {ascii_value}")
print(f"    SEED_NUM (B): {seed_num}")
print(f"    Total Loops (C): {loops}")
print(f"    Execution Log: Calculating ({ascii_value} × {seed_num}) - {loops}")
print(f"    Verification Hash: {verification_hash}")
print("\nAssessment Data:")
print(f"    First Letter ASCII (A): {ascii_value}")
print(f"    SEED_NUM (B): {seed_num}")
print(f"    Total Loops (C): {loops}")
print(f"    Verification Hash: {verification_hash}")
```

```python
# Initialize variables
correct_password = "******"
attempt_counter = 0


print("EXERCISE 3: SECURE INPUT LOOP")
print(f"Correct Password Set: {correct_password}")
print("Execution Log:")
print("System locked. Enter password to unlock.")

while True:
    try:
        pwd = input("Enter password: ")
        attempt_counter += 1               #this code is from AI
        print(f"  \nAttempt #{attempt_counter}: Input received")
        
        if pwd == correct_password:        #this code is from AI
            print("  Access Granted! System unlocked.")
            break
        else:
            print("  Access Denied. Try again.")
    except KeyboardInterrupt:
        print("\n  System shutdown requested.")
        break

print(f"\nAssessment Data:")
print(f"    Correct Password Set: {correct_password}")        #this code is from AI
print(f"    Input Attempts: {attempt_counter}")
print(f"    Attempts Before Success: {'N/A - Failed' if pwd != correct_password else attempt_counter}")
```
